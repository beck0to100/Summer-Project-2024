# Model Building, Model Training, & Model Evaluation


```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.layers import Input, Embedding, Dense, Flatten, LayerNormalization, Dropout, Concatenate
from tensorflow.keras.models import Model

# Load the preprocessed CSV file
updated_data_with_mean50_path = '/content/drive/MyDrive/UGA_files/Research_2024_05_29/TN_seq_dataset/updated_data_with_mean50.csv'
updated_data_with_mean50 = pd.read_csv(updated_data_with_mean50_path)

# Tokenization for 'Gene' column
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(updated_data_with_mean50['Gene'])
sequences = tokenizer.texts_to_sequences(updated_data_with_mean50['Gene'])

# Padding sequences
max_sequence_length = 10  # Adjust this value as needed
padded_sequences = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_sequence_length, padding='post')

# Create attention masks
attention_masks = np.where(padded_sequences != 0, 1, 0)

# Create DataFrame for padded sequences
X_text = pd.DataFrame(padded_sequences, columns=[f'Gene_{i}' for i in range(max_sequence_length)])

# Identify and encode non-numeric columns (excluding 'Gene')
non_numeric_columns = updated_data_with_mean50.select_dtypes(include=['object']).columns.difference(['Gene'])
if len(non_numeric_columns) > 0:
    encoder = OneHotEncoder(sparse_output=False)
    encoded_non_numeric = encoder.fit_transform(updated_data_with_mean50[non_numeric_columns])
    encoded_non_numeric_df = pd.DataFrame(encoded_non_numeric, columns=encoder.get_feature_names_out(non_numeric_columns))
    updated_data_with_mean50 = updated_data_with_mean50.drop(columns=non_numeric_columns)
    updated_data_with_mean50 = pd.concat([updated_data_with_mean50, encoded_non_numeric_df], axis=1)

# Select numeric columns and combine with encoded non-numeric data
numeric_columns = updated_data_with_mean50.select_dtypes(include=[np.number]).columns.difference(['Mean50_DMSO_5'])
X_other = updated_data_with_mean50[numeric_columns]

# Prepare final input and output
X_text = X_text.values.astype(np.float32)
X_other = X_other.values.astype(np.float32)
y = updated_data_with_mean50['Mean50_DMSO_5'].values.astype(np.float32)

# Convert to TensorFlow dataset
dataset = tf.data.Dataset.from_tensor_slices(({"input_ids": X_text, "attention_mask": attention_masks, "other_features": X_other}, y))

# Split into training, validation, and test sets
train_size = int(0.6 * len(dataset))
val_size = int(0.2 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size).take(val_size)
test_dataset = dataset.skip(train_size + val_size)

# Batch the datasets
batch_size = 256  # Adjust as needed
train_dataset = train_dataset.batch(batch_size)
val_dataset = val_dataset.batch(batch_size)
test_dataset = test_dataset.batch(batch_size)

# Prefetch the datasets for performance optimization
train_dataset = train_dataset.prefetch(tf.data.AUTOTUNE)
val_dataset = val_dataset.prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.prefetch(tf.data.AUTOTUNE)

# Define the positional encoding layer
class PositionalEncoding(tf.keras.layers.Layer):
    def __init__(self, position, d_model):
        super(PositionalEncoding, self).__init__()
        self.pos_encoding = self.positional_encoding(position, d_model)

    def get_config(self):
        config = super().get_config().copy()
        return config

    def positional_encoding(self, position, d_model):
        angle_rads = self.get_angles(np.arange(position)[:, np.newaxis], np.arange(d_model)[np.newaxis, :], d_model)
        angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
        angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])
        pos_encoding = angle_rads[np.newaxis, ...]
        return tf.cast(pos_encoding, dtype=tf.float32)

    def get_angles(self, pos, i, d_model):
        angle_rates = 1 / np.power(10000, (2 * (i // 2)) / np.float32(d_model))
        return pos * angle_rates

    def call(self, inputs):
        return inputs + self.pos_encoding[:, :tf.shape(inputs)[1], :]

# Transformer Encoder Layer
def transformer_encoder_layer(units, d_model, num_heads, dropout, name):
    inputs = tf.keras.Input(shape=(None, d_model), name=f"{name}_inputs")
    padding_mask = tf.keras.Input(shape=(None,), name=f"{name}_padding_mask")

    attention = tf.keras.layers.MultiHeadAttention(num_heads=num_heads, key_dim=d_model, name=f"{name}_attention")(
        query=inputs, key=inputs, value=inputs, attention_mask=padding_mask[:, tf.newaxis, tf.newaxis, :])
    attention = tf.keras.layers.Dropout(rate=dropout)(attention)
    attention = tf.keras.layers.LayerNormalization(epsilon=1e-6)(inputs + attention)

    outputs = tf.keras.layers.Dense(units=units, activation='relu', name=f"{name}_dense_1")(attention)
    outputs = tf.keras.layers.Dense(units=d_model, name=f"{name}_dense_2")(outputs)
    outputs = tf.keras.layers.Dropout(rate=dropout)(outputs)
    outputs = tf.keras.layers.LayerNormalization(epsilon=1e-6)(attention + outputs)

    return tf.keras.Model(inputs=[inputs, padding_mask], outputs=outputs, name=name)

# Define the Transformer model
def build_transformer_model(input_shape, vocab_size, num_layers, units, d_model, num_heads, dropout, max_seq_len, num_other_features):
    input_ids = tf.keras.Input(shape=input_shape, dtype=tf.int32, name="input_ids")
    attention_mask = tf.keras.Input(shape=(input_shape[0],), dtype=tf.float32, name="attention_mask")
    other_features = tf.keras.Input(shape=(num_other_features,), dtype=tf.float32, name="other_features")

    # Embedding layer
    embedding_layer = Embedding(input_dim=vocab_size, output_dim=d_model, name="embedding")(input_ids)

    # Positional Encoding
    pos_encoding = PositionalEncoding(max_seq_len, d_model)(embedding_layer)

    # Transformer Encoder Layers
    enc_output = pos_encoding
    for i in range(num_layers):
        enc_output = transformer_encoder_layer(units=units, d_model=d_model, num_heads=num_heads, dropout=dropout, name=f"encoder_layer_{i}")(
            [enc_output, attention_mask])

    # Flatten the output
    enc_output = Flatten(name="flatten")(enc_output)

    # Concatenate transformer output with other features
    concatenated = Concatenate(name="concatenate")([enc_output, other_features])

    # Dense layers
    dense = Dense(units=units, activation='relu', name="dense_1")(concatenated)
    dense = Dropout(rate=dropout, name="dropout_1")(dense)

    # Output layer
    outputs = Dense(units=1, activation='linear', name="outputs")(dense)

    return Model(inputs=[input_ids, attention_mask, other_features], outputs=outputs, name="transformer")

# Parameters for the model
input_shape = (max_sequence_length,)  # Adjust based on the actual input shape
vocab_size = 20000  # Adjust based on the actual vocabulary size
num_layers = 2
units = 128
d_model = 64
num_heads = 4
dropout = 0.1
max_seq_len = max_sequence_length
num_other_features = X_other.shape[1]

# Build the model
transformer_model = build_transformer_model(input_shape, vocab_size, num_layers, units, d_model, num_heads, dropout, max_seq_len, num_other_features)

# Compile the model
transformer_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Display the model summary
transformer_model.summary()

# Callbacks for early stopping and model checkpointing
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',  # Monitor validation loss
    patience=3,          # Stop after 3 epochs with no improvement
    restore_best_weights=True  # Restore model to best weights observed during training
)

model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath='/content/drive/MyDrive/UGA_files/Research_2024_05_29/TN_seq_dataset/new_best_model.h5',  # Filepath to save the model
    monitor='val_loss',  # Monitor validation loss
    save_best_only=True,  # Save only the best model
    save_weights_only=False,  # Save the full model (architecture + weights)
    mode='min',  # The model is considered better when the validation loss is minimized
    verbose=1  # Print messages when the model is saved
)

# Train the model with the callbacks
history = transformer_model.fit(
    train_dataset,  # Training data
    validation_data=val_dataset,  # Validation data
    epochs=20,  # Number of epochs
    callbacks=[early_stopping, model_checkpoint]  # List of callbacks
)

# Evaluate the model
test_loss, test_mae = transformer_model.evaluate(test_dataset)
print(f'Test Loss: {test_loss}')
print(f'Test MAE: {test_mae}')

# Collect evaluation metrics
evaluation_results = {
    'Metric': ['Test Loss', 'Test MAE'],
    'Value': [test_loss, test_mae]
}
evaluation_df = pd.DataFrame(evaluation_results)

# Save the metrics to a CSV file
evaluation_df.to_csv('/content/drive/MyDrive/UGA_files/Research_2024_05_29/TN_seq_dataset/new_evaluation_results.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import shap
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Input, Embedding, Dense, Flatten, LayerNormalization, Dropout, Concatenate
from tensorflow.keras.models import Model
from sklearn.preprocessing import OneHotEncoder

# Load the trained model
model_path = '/content/drive/MyDrive/UGA_files/Research_2024_05_29/TN_seq_dataset/new_best_model.h5'
transformer_model = load_model(model_path)

# Load preprocessed data
updated_data_with_mean50_path = '/content/drive/MyDrive/UGA_files/Research_2024_05_29/TN_seq_dataset/updated_data_with_mean50.csv'
updated_data_with_mean50 = pd.read_csv(updated_data_with_mean50_path)

# Tokenization for 'Gene' column
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(updated_data_with_mean50['Gene'])
sequences = tokenizer.texts_to_sequences(updated_data_with_mean50['Gene'])
max_sequence_length = 10  # Adjust based on your model
padded_sequences = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_sequence_length, padding='post')

# Create attention masks
attention_masks = np.where(padded_sequences != 0, 1, 0)

# Features for SHAP
X_text = padded_sequences.astype(np.float32)
X_other = updated_data_with_mean50.drop(columns=['Gene', 'Mean50_DMSO_5']).values.astype(np.float32)

# Define a SHAP explainer
explainer = shap.Explainer(transformer_model, masker=shap.maskers.Text(tokenizer))

# Compute SHAP values
shap_values = explainer(X_text, attention_mask=attention_masks, other_features=X_other)

# Summarize the effects of all the features
shap.summary_plot(shap_values, X_text)
```
